**Q1: Write a program that takes a word as an input and print the number of vowels in the word.**

```python
word = input("Enter a word: ")
vowels = "aeiouAEIOU"
count = 0
for letter in word:
    if letter in vowels:
        count += 1
print("Number of vowels:", count)
```

**Q2: Iterate through the following list of animals and print each one in all caps.
animals=['tiger', 'elephant', 'monkey', 'zebra', 'panther']**

```python
animals = ['tiger', 'elephant', 'monkey', 'zebra', 'panther']
for animal in animals:
    print(animal.upper())
```

**Q3: Write a program that iterates from 1 to 20, printing each number and whether it's odd or even.**

```python
for i in range(1, 21):
    if i % 2 == 0:
        print(i, "is even")
    else:
        print(i, "is odd")
```

**Q4: Write a program to check if a string is a palindrome or not.**

```python
text = input("Enter a string: ")
if text == text[::-1]:
    print("Palindrome")
else:
    print("Not a palindrome")
```

**Q5: Write a function sum_of_integers(a, b) that takes two integers as input from the user and returns their sum.**

```python
def sum_of_integers(a, b):
    return a + b

a = int(input("Enter first integer: "))
b = int(input("Enter second integer: "))
print("Sum:", sum_of_integers(a, b))
```
